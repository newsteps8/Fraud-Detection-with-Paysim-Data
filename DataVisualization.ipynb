```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import OrdinalEncoder
```

```python
df = pd.read_csv('PS_20174392719_1491204439457_log.csv')
df.info()
```

```python
feature_list = ["step", "amount", "oldbalanceOrg",
                "newbalanceOrig", "oldbalanceDest",
                "newbalanceDest","isFraud", "isFlaggedFraud"]

count = 1
plt.figure(figsize=(20,20))
for i in feature_list:
    plt.subplot(3, 3, count)
    sns.distplot(df[i],bins=10,kde_kws={'bw':0.1})
    count += 1
```

```python
feature_list = ["nameOrig", "nameDest"]

ord_enc = OrdinalEncoder()
df["nameOrig"] = ord_enc.fit_transform(df[["nameOrig"]])
df["nameDest"] = ord_enc.fit_transform(df[["nameDest"]])

count = 1
plt.figure(figsize=(20,20))
for i in feature_list:
    plt.subplot(3, 3, count)
    sns.distplot(df[i],bins=10,kde_kws={'bw':0.1})
    count += 1
```

```python
counts_list = df['type'].value_counts()
values_list = counts_list.index.tolist()
plt.bar(values_list, counts_list, align= 'center')
```
